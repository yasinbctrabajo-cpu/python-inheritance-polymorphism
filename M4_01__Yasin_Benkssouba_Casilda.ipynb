**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



````python
# Creo la clase Vehiculo con un constructor que recibe la marca y el modelo del vehículo
class Vehiculos:
    def __init__ (self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
# Creo el método repostar que imprime un mensaje indicando que el vehículo necesita repostar gasolina
    def repostar(self):
        print("Este vehiculo tiene que repostar gasolina")

# Creo la clase VElectricos que hereda de la clase Vehiculos y tiene un constructor que recibe la marca, el modelo y la capacidad de la autonomía
class VElectricos(Vehiculos):
    def __init__ (self, marca, modelo, autonomia):
        super().__init__(marca, modelo)
        self.autonomia = autonomia
    def repostar(self):
        print("Este vehiculo tiene que repostar electricidad")

# Creo una clase BicicletaElectrica que herede de la clase vElectricos y Vehiculos, dando prioridad a VElectricos
class BicicletaElectrica(VElectricos, Vehiculos):
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo, autonomia)
# Creo la clase Quad que hereda de VElectricos y Vehiculos, aunque hereda de VElectricos (que ya hereda de Vehiculos)
class Quad(VElectricos, Vehiculos):
    def __init__ (self, marca, modelo, autonomia):
        super().__init__(marca, modelo, autonomia)
# sobrescribo el método repostar() (override) para dar prioridad al comportamiento de Vehiculos (gasolina).
    def repostar(self): 
        print("Este vehiculo tiene que repostar gasolina")

# Creo un objeto de BicicletaElectrica y otro de Quad
b = BicicletaElectrica("Santa Cruz", "Vala 90", 90)
q = Quad("Yamaha", "Banshee 350", 73)

flota = [b, q]
for vehiculo in flota:
    vehiculo.repostar()

        
````

Si puedes, incluye alguna técnica de polimorfismo.**

````python
# La temática será sobre un Videojuego de guerra, donde habrá diferentes tipos de personajes (Personaje(superclase), Guerrero, Jinete, Mago, Arquero, Enemigo) y cada una tendrá un método para atacar y otro para defender.

# Constructor base con los atributos comunes a todos los personajes
class Personaje:
    def __init__(self, nombre, nivel, salud):
        self.nombre = nombre
        self.nivel = nivel
        self.salud = salud

    # Método genérico de ataque y defensa que será sobrescrito por las clases hijas
    def atacar(self):
        print(f"{self.nombre} ataca con su arma")

    def defender(self):
        print(f"{self.nombre} se defiende con su escudo")

# Clase intermedia que hereda de Personaje (herencia multinivel) y que será la base para los personajes jugadores (Guerrero, Jinete, Mago, Arquero)
class Jugador(Personaje):
    def __init__(self, nombre, nivel, salud, arma):
        super().__init__(nombre, nivel, salud)
        self.arma = arma

# Se sobrescribe (override) el método atacar para personalizarlo
    def atacar(self):
        print(f"{self.nombre} ataca con su {self.arma}")

# También se sobrescribe defender para diferenciarlo del base
    def defender(self):
        print(f"{self.nombre} se defiende con su armadura")

# Clases específicas que heredan de Jugador, cada una define su arma concreta llamando al constructor del padre
class Guerrero(Jugador):
    def __init__(self, nombre, nivel, salud):
        super().__init__(nombre, nivel, salud, "Martillo")

class Jinete(Jugador):
    def __init__ (self, nombre, nivel, salud):
        super().__init__(nombre, nivel, salud, "Lanza")

class Mago(Jugador):
    def __init__(self, nombre, nivel, salud):
        super().__init__(nombre, nivel, salud, "Báculo")

class Arquero(Jugador):
    def __init__(self, nombre, nivel, salud):
        super().__init__(nombre, nivel, salud, "Arco")

# Clase Enemigo que hereda directamente de Personaje
class Enemigo(Personaje):
    def __init__(self, nombre, nivel, salud, tipo):
        super().__init__(nombre, nivel, salud)
        self.tipo = tipo

# Override para personalizar comportamiento de ataque y defensa del enemigo
    def atacar(self):
        print(f"{self.nombre} ataca con su {self.tipo}")

    def defender(self):
        print(f"{self.nombre} se defiende con su {self.tipo}")


# Creo un objeto de cada clase y muestro sus ataques y defensas
guerrero = Guerrero("Thor", 10, 100)
jinete = Jinete("Centauro", 8, 80)
mago = Mago("Merlín", 12, 70)
arquero = Arquero("Legolas", 9, 90)
enemigo = Enemigo("Orco", 5, 50, "Hacha")
personajes = [guerrero, jinete, mago, arquero, enemigo]
for i, personaje in enumerate(personajes):
    if i > 0:
        print("-" * 20)
    personaje.atacar()
    personaje.defender()


         
````
